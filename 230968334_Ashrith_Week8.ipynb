```python
1) Rule-Based Logical Agent for Smart Lighting
A smart building uses sensors to control lighting automatically. The system must decide
whether to turn lights ON or OFF based on logical rules.
The system perceives:
• MotionDetected
• IsDark
• ManualOverride
Represent the following knowledge using propositional logic:
• If it is dark and motion is detected, turn the light ON
• If there is no motion, turn the light OFF
• If manual override is ON, ignore all automatic rules
Design a logical agent that:
• Takes sensor readings as percepts
• Updates its knowledge base
• Infers the correct action
Given a set of percepts, determine the agent’s action using logical inference.
```

```python
def smart_lighting(percepts):
    MotionDetected = percepts.get("MotionDetected", False)
    IsDark = percepts.get("IsDark", False)
    ManualOverride = percepts.get("ManualOverride", False)

    if ManualOverride:
        return "Manual Control (Ignore Rules)"

    if IsDark and MotionDetected:
        return "Light ON"
    elif not MotionDetected:
        return "Light OFF"
    else:
        return "No Change"


# Test
print(smart_lighting({
    "MotionDetected": True,
    "IsDark": True,
    "ManualOverride": False
}))
```

2) Medical Diagnosis Logical Agent
A medical decision-support system assists doctors by reasoning over symptoms to
suggest possible conditions.
Given symptoms:
• Fever
• Cough
• ChestPain
Diseases:
• Flu
• Pneumonia
Represent the following medical knowledge using propositional logic:
• Fever ∧ Cough → Flu
• Fever ∧ Cough ∧ ChestPain → Pneumonia
• Pneumonia → HospitalizationRequired
Build a knowledge base and use forward chaining to infer:
• Possible diseases
• Required actions
Explain how this agent differs from a machine-learning classifier.

```python
def medical_agent(symptoms):
    facts = set(symptoms)
    inferred = set()

    # Rule 1
    if "Fever" in facts and "Cough" in facts:
        inferred.add("Flu")

    # Rule 2
    if "Fever" in facts and "Cough" in facts and "ChestPain" in facts:
        inferred.add("Pneumonia")

    # Rule 3
    if "Pneumonia" in inferred:
        inferred.add("HospitalizationRequired")

    return inferred


# Test
print(medical_agent({"Fever", "Cough", "ChestPain"}))
```

```python
3) Logical Agent for Traffic Violation Detection
A smart traffic monitoring system detects violations using logical rules rather than imagebased ML models.
Percepts include:
• SpeedAboveLimit
• SignalRed
• VehicleCrossedLine
• EmergencyVehicle
Define logical rules to infer:
• SpeedViolation
• SignalViolation
• Exceptions for emergency vehicles
Implement a logical agent cycle:
• Perceive
• Update knowledge base
• Infer violations
• Decide enforcement action
Given a sequence of percepts over time, show how conclusions change.
```

```python
def traffic_agent(percepts):
    if percepts.get("EmergencyVehicle"):
        return "No Violation (Emergency)"

    violations = []

    if percepts.get("SpeedAboveLimit"):
        violations.append("SpeedViolation")

    if percepts.get("SignalRed") and percepts.get("VehicleCrossedLine"):
        violations.append("SignalViolation")

    return violations if violations else ["No Violation"]


# Test sequence
data = [
    {"SpeedAboveLimit": True},
    {"SignalRed": True, "VehicleCrossedLine": True},
    {"EmergencyVehicle": True}
]

for d in data:
    print(traffic_agent(d))
```

```python
4) Logical Agent for Cybersecurity Intrusion Detection
An enterprise network uses a logical agent to detect potential cyber intrusions based on
system events.
Percepts include:
• MultipleFailedLogins
• LoginFromUnknownIP
• AccessToSensitiveFiles
• AdminPrivileges
Encode security knowledge using propositional logic:
• MultipleFailedLogins ∧ LoginFromUnknownIP → SuspiciousLogin
• SuspiciousLogin ∧ AdminPrivileges → HighRiskIntrusion
• HighRiskIntrusion → TriggerAlert
Design a knowledge-based agent that:
• Continuously updates facts
• Infers threat levels
• Decides defensive actions
```

```python
def cybersecurity_agent(percepts):
    facts = set(percepts)
    inferred = set()

    if "MultipleFailedLogins" in facts and "LoginFromUnknownIP" in facts:
        inferred.add("SuspiciousLogin")

    if "SuspiciousLogin" in inferred and "AdminPrivileges" in facts:
        inferred.add("HighRiskIntrusion")

    if "HighRiskIntrusion" in inferred:
        inferred.add("TriggerAlert")

    return inferred


# Test
print(cybersecurity_agent({
    "MultipleFailedLogins",
    "LoginFromUnknownIP",
    "AdminPrivileges"
}))
```

```python
5) Autonomous Warehouse Robot Using Logical Reasoning
An autonomous warehouse robot must decide how to move and operate safely using
logical reasoning rather than learning.
Percepts:
• ObstacleAhead
• PathClear
• CarryingFragileItem
• HumanNearby
• DestinationReached
Represent operational rules using propositional logic:
• ObstacleAhead → Stop
• CarryingFragileItem ∧ ObstacleAhead → SlowDown
• HumanNearby → ReduceSpeed
• PathClear ∧ ¬HumanNearby → MoveForward
Design a logical agent architecture showing:
• Knowledge base
• Inference engine
• Action selection
```

```python
def warehouse_robot(percepts):
    ObstacleAhead = percepts.get("ObstacleAhead", False)
    PathClear = percepts.get("PathClear", False)
    CarryingFragileItem = percepts.get("CarryingFragileItem", False)
    HumanNearby = percepts.get("HumanNearby", False)

    if ObstacleAhead and CarryingFragileItem:
        return "SlowDown"
    if ObstacleAhead:
        return "Stop"
    if HumanNearby:
        return "ReduceSpeed"
    if PathClear and not HumanNearby:
        return "MoveForward"

    return "Idle"


# Test
print(warehouse_robot({
    "ObstacleAhead": True,
    "CarryingFragileItem": True
}))
```
