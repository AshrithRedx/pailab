```python
1. Modern cybersecurity systems often rely on rule-based reasoning to detect threats based
on system activity. Unlike machine learning models, these systems provide clear,
explainable decisions using logical inference.
Design a logical inference engine that:
• Accepts facts and rules
• Converts them into Conjunctive Normal Form (CNF)
• Applies resolution to determine whether a security threat exists
Design a Cybersecurity Threat Detection Agent that determines whether a system intrusion
has occurred using propositional logic and resolution.
The system receives Facts (observed events) and Rules (security knowledge)
The agent must:
1. Convert all rules into CNF
2. Apply resolution-based theorem proving
3. Determine whether:
IntrusionDetected = TRUE or FALSE
Input: Facts (Percepts)
Examples:
• MultipleFailedLogins
• LoginFromUnknownIP
• AccessToSensitiveFiles
• AdminPrivileges
Input: Rules (Knowledge Base)
Examples:
MultipleFailedLogins ∧ LoginFromUnknownIP → SuspiciousLogin
SuspiciousLogin ∧ AdminPrivileges → HighRiskIntrusion
HighRiskIntrusion → IntrusionDetected
AccessToSensitiveFiles ∧ ¬AdminPrivileges → IntrusionDetected
Knowledge Representation
1. Represent all rules using propositional logic
2. Define:
• Atomic propositions
• Implications
• Negations 
Part B: CNF Conversion
Convert all rules into Conjunctive Normal Form (CNF):
Steps to follow:
1. Eliminate implications
2. Move negations inward
3. Standardize expressions
4. Convert to conjunction of disjunctions
Resolution Algorithm Implementation
1. Add negation of query:
¬IntrusionDetected
2. Apply resolution rule:
(A ∨ B), (¬B ∨ C) ⇒ (A ∨ C)
3. Continue until:
• Empty clause is derived → TRUE (Intrusion detected)
• No new clauses → FALSE
Program must:
Input:
• List of facts
• List of rules
Output:
• CNF clauses
• Resolution steps (intermediate clauses)
• Final conclusion:
IntrusionDetected: TRUE / FALSE
Scenario Testing
Test your system on at least 3 scenarios:
Scenario 1: Low Risk
MultipleFailedLogins
Expected: No intrusion
Scenario 2: Suspicious Activity
MultipleFailedLogins
LoginFromUnknownIP
Expected: Suspicious but not confirmed intrusion
Scenario 3: High Risk Intrusion
MultipleFailedLogins
LoginFromUnknownIP
AdminPrivileges
Expected: IntrusionDetected = TRUE
Output Format Example
CNF Clauses:
1. ¬MultipleFailedLogins ∨ ¬LoginFromUnknownIP ∨ SuspiciousLogin
2. ¬SuspiciousLogin ∨ ¬AdminPrivileges ∨ HighRiskIntrusion
3. ¬HighRiskIntrusion ∨ IntrusionDetected
Resolution Steps:
Derived: IntrusionDetected
Final Output: IntrusionDetected = TRUE
```

```python
from itertools import combinations

# ==============================
# HELPER FUNCTIONS
# ==============================

def negate(literal):
    return literal[1:] if literal.startswith("¬") else "¬" + literal

def resolve(ci, cj):
    resolvents = []
    for li in ci:
        for lj in cj:
            if li == negate(lj):
                new_clause = set(ci + cj)
                new_clause.discard(li)
                new_clause.discard(lj)
                resolvents.append(list(new_clause))
    return resolvents

# ==============================
# RESOLUTION ALGORITHM
# ==============================

def resolution(kb, query):
    clauses = kb.copy()
    clauses.append([negate(query)])

    print("\nInitial Clauses:")
    for c in clauses:
        print(c)

    new = []

    while True:
        pairs = list(combinations(clauses, 2))

        for (ci, cj) in pairs:
            resolvents = resolve(ci, cj)
            for r in resolvents:
                print("Derived:", r)
                if r == []:
                    return True
                new.append(r)

        if all(n in clauses for n in new):
            return False

        for n in new:
            if n not in clauses:
                clauses.append(n)
```

```python
kb = [
    ["¬MultipleFailedLogins", "¬LoginFromUnknownIP", "SuspiciousLogin"],
    ["¬SuspiciousLogin", "¬AdminPrivileges", "HighRiskIntrusion"],
    ["¬HighRiskIntrusion", "IntrusionDetected"],
    ["¬AccessToSensitiveFiles", "AdminPrivileges", "IntrusionDetected"]
]
```

```python
def test_scenario(facts, name):
    print("\n===== ", name, " =====")

    clauses = kb.copy()

    # Add facts as unit clauses
    for f in facts:
        clauses.append([f])

    result = resolution(clauses, "IntrusionDetected")

    print("\nFinal Output:")
    print("IntrusionDetected =", "TRUE" if result else "FALSE")


# Scenario 1: Low Risk
test_scenario(
    ["MultipleFailedLogins"],
    "Scenario 1: Low Risk"
)

# Scenario 2: Suspicious
test_scenario(
    ["MultipleFailedLogins", "LoginFromUnknownIP"],
    "Scenario 2: Suspicious Activity"
)

# Scenario 3: High Risk
test_scenario(
    ["MultipleFailedLogins", "LoginFromUnknownIP", "AdminPrivileges"],
    "Scenario 3: High Risk"
)
```

```python
from itertools import combinations

def negate(literal):
    return literal[1:] if literal.startswith("¬") else "¬" + literal

def resolve(ci, cj):
    resolvents = []
    for li in ci:
        for lj in cj:
            if li == negate(lj):
                new_clause = set(ci + cj)
                new_clause.discard(li)
                new_clause.discard(lj)
                resolvents.append(list(new_clause))
    return resolvents

def resolution(kb, query):
    clauses = kb.copy()
    clauses.append([negate(query)])

    print("\nCNF Clauses:")
    for c in clauses:
        print(c)

    new = []

    while True:
        pairs = list(combinations(clauses, 2))

        for (ci, cj) in pairs:
            resolvents = resolve(ci, cj)
            for r in resolvents:
                print("Derived:", r)
                if r == []:
                    return True
                new.append(r)

        if all(n in clauses for n in new):
            return False

        for n in new:
            if n not in clauses:
                clauses.append(n)
```
