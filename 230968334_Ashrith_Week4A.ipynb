<b>Experiment 2:</b> Greedy Best-First Search for Maze Navigation<br>
Problem Statement:<br>
Consider a maze represented as a graph, where each node corresponds to a location in the maze<br>
and edges represent valid paths between locations.<br>
An agent is placed at a start node and must reach a goal node by traversing the maze.<br>
Unlike BFS, the agent will choose the next node to explore based on a heuristic estimate of how
close it is to the goal.<br>
Task:<br>
1. Formulate the maze navigation problem as a search problem by clearly defining: State space,<br>
Initial state, Actions, Transition model, Goal test, Path cost, Heuristic function (ℎ(𝑛))<br>
2. Implement the Greedy Best-First Search algorithm to find a path from the start node to the<br>
goal node.<br>
3. Use the heuristic function:<br>
ℎ(𝑛)=Manhattan distance between node 𝑛 and the goal<br>
4. Print the path returned by Greedy Best-First Search.<br>
5. Count and display the number of nodes expanded during the search.<br>
Constraints:<br>

The maze is unweighted.<br>
The environment is fully observable, deterministic, static, discrete, and single-agent.<br>
Greedy Best-First Search must be implemented without using built-in graph search libraries.<br>
The algorithm should expand the node with the lowest heuristic value first.<br>

```python
import heapq

def manhattan_distance(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def greedy_best_first_search(maze, start, goal):
    rows, cols = len(maze), len(maze[0])

    frontier = []
    heapq.heappush(frontier, (manhattan_distance(start, goal), start))

    came_from = {start:None}
    visited = set()
    nodes_expanded = 0

    while frontier:
        _, current = heapq.heappop(frontier)

        if current in visited:
            continue

        visited.add(current)
        nodes_expanded += 1

        if current == goal:
            break

        r, c = current
        neighbors = [
            (r - 1, c),  # up
            (r + 1, c),  # down
            (r, c - 1),  # left
            (r, c + 1)   # right
        ]

        for nr, nc in neighbors:
            if (0 <= nr < rows and
                0 <= nc < cols and
                maze[nr][nc] == 0 and
                (nr, nc) not in visited):

                neighbor = (nr, nc)
                if neighbor not in came_from:
                    came_from[neighbor] = current
                    h = manhattan_distance(neighbor, goal)
                    heapq.heappush(frontier, (h, neighbor))

    # Reconstruct path
    path = []
    if goal in came_from:
        node = goal
        while node:
            path.append(node)
            node = came_from[node]
        path.reverse()

    return path, nodes_expanded
        
```

```python
# 0 = free space, 1 = wall
maze = [
    [0, 0, 0, 0],
    [1, 1, 0, 1],
    [0, 0, 0, 0],
    [0, 1, 1, 0]
]

start = (0, 0)
goal = (3, 0)

path, expanded = greedy_best_first_search(maze, start, goal)

print("Path found:", path)
print("Number of nodes expanded:", expanded)
```

<b>Experiment 3:</b> A* Search for Optimal Maze Navigation<br>
Problem Statement:<br>
Consider a maze represented as a graph, where each node corresponds to a location in the<br> maze<br>
and edges represent valid paths between locations.<br>
An agent is placed at a start node and must reach a goal node.<br>
To ensure optimal navigation, the agent will use the A* search algorithm, which combines:<br>
Actual cost from start 𝑔(𝑛) and Heuristic estimate to goal ℎ(𝑛)<br>
Task:1. Formulate the maze navigation problem as a search problem by clearly defining:<br>State space,
Initial state, Actions, Transition model, Goal test, Path cost, Heuristic function ℎ(𝑛)<br>
2. Implement the A* Search algorithm to find the optimal shortest path from the start node
to the goal node.<br>
3. Use the evaluation function: 𝑓(𝑛)=𝑔(𝑛)+ℎ(𝑛)
where:<br>
𝑔(𝑛) = path cost from start to node 𝑛<br>
ℎ(𝑛) = Manhattan distance from node 𝑛 to goal<br>
4. Print the path returned by A* Search.<br>
5. Count and display the number of nodes expanded during the search.<br>
Constraints:<br>

Each move has uniform cost = 1.<br>
The heuristic must be admissible (never overestimates).<br>
The environment is fully observable, deterministic, static, discrete, and single-agent.<br>
A* must be implemented without using any built-in graph search libraries.<br>
The algorithm should always expand the node with the lowest 𝑓(𝑛)<br>

```python
import heapq

def manhattan_distance(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def a_star_search(maze, start, goal):
    rows, cols = len(maze), len(maze[0])

    frontier = []
    heapq.heappush(frontier, (0, 0, start))

    came_from = {start: None}
    g_cost = {start: 0}
    visited = set()

    nodes_expanded = 0

    while frontier:
        _, current_g, current = heapq.heappop(frontier)

        if current in visited:
            continue

        visited.add(current)
        nodes_expanded += 1

        # Goal test
        if current == goal:
            break

        r, c = current
        neighbors = [
            (r - 1, c),  # up
            (r + 1, c),  # down
            (r, c - 1),  # left
            (r, c + 1)   # right
        ]

        for nr, nc in neighbors:
            if 0 <= nr < rows and 0 <= nc < cols and maze[nr][nc] == 0:
                neighbor = (nr, nc)
                tentative_g = current_g + 1

                if neighbor not in g_cost or tentative_g < g_cost[neighbor]:
                    g_cost[neighbor] = tentative_g
                    h = manhattan_distance(neighbor, goal)
                    f = tentative_g + h

                    came_from[neighbor] = current
                    heapq.heappush(frontier, (f, tentative_g, neighbor))

    path = []
    if goal in came_from:
        node = goal
        while node is not None:
            path.append(node)
            node = came_from[node]
        path.reverse()

    return path, nodes_expanded
```

```python
maze = [
    [0, 0, 0, 0],
    [1, 1, 0, 1],
    [0, 0, 0, 0],
    [0, 1, 1, 0]
]

start = (0, 0)
goal = (3, 0)

path, expanded = a_star_search(maze, start, goal)

print("Optimal path found:", path)
print("Number of nodes expanded:", expanded)
```
