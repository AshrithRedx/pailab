```python
Lab Exercise 1: Tic-Tac-Toe AI using Minimax and Alpha–Beta
Objective: Implement a Tic-Tac-Toe AI agent that plays optimally using: Minimax, Alpha–Beta
pruning
Tic-Tac-Toe is a two-player, zero-sum, deterministic, turn-based game with perfect information,
making it an ideal testbed for understanding adversarial search algorithms.
In this lab, you will design an intelligent game-playing agent for Tic-Tac-Toe using:
• Minimax algorithm (optimal decision-making)
• Alpha–Beta pruning (search optimization)
Design an AI agent that plays Tic-Tac-Toe optimally against a human or another AI agent.
The agent must:
• Analyze the game state
• Anticipate the opponent’s best possible moves
• Select actions that maximize its chance of winning while minimizing losses
Constraints for Implementation:
• Board size: 3×3
• Symbols:
• AI = X (MAX)
• Opponent = O (MIN) (either human input or another AI)
• Start player: X goes first
• Assume both players play optimally (for Minimax correctness).
• Players alternate turns
• The game ends when One player gets three marks in a row (horizontal, vertical, or diagonal),
or The board is full (draw)
Formulate Tic-Tac-Toe as an adversarial search problem by defining:
Part A: Core Functions (Must Implement)
1. State representation
Use either: list of length 9 (recommended): [' ', ' ', ...] OR 3×3 list
2. Initial State: Empty board
3. print_board(state): Displays the board clearly.
4. actions(state) :Returns list of all empty positions.
5. Actions : Placing a mark (X or O) in any empty cell
6. result(state, action, player)
Returns a new state after placing player mark at action.
7. terminal(state) : Returns True if win/loss/draw.
8. utility(state)
Return: +1 if X wins, -1 if O wins, 0 draw / non-terminal should not be evaluated here (or
return None)
9. player(state) : Returns whose turn it is (based on count of X and O).
Part B: Minimax Implementation (Must)
Implement recursion:
• minimax_decision(state) → returns best move for current player (or for X only)
• max_value(state)
• min_value(state)
3. Return:
• Optimal move for MAX
• Minimax value of the current board
Node counting requirement: Maintain a global/local counter
• nodes_minimax += 1 every time you evaluate a node/state.
Part C: Alpha–Beta Implementation (Must)
Modify the Minimax implementation to include:
Alpha (α): best value found so far for MAX
Beta (β): best value found so far for MIN
Apply pruning when: β≤α
Ensure: Final decision remains the same as Minimax
Fewer nodes are evaluated: Maintain: nodes_ab += 1
Program Execution Modes (Choose one; recommend Mode 1)
Mode 1 : Human vs AI
Human plays O
AI plays X using Minimax / AlphaBeta
Mode 2 : AI vs AI
X uses Minimax
O uses AlphaBeta (or both AlphaBeta)
Output Requirements
Program must print after every AI move:
a. Current board
b. AI selected move (row, col or index 0–8)
c. Game status: “X wins / O wins / Draw / Continue”
d. At the end, print:
Comparison
Nodes evaluated using Minimax: _____
Nodes evaluated using Alpha–Beta: _____
```

```python
import math

# ==============================
# GLOBAL NODE COUNTS
# ==============================
nodes_minimax = 0
nodes_ab = 0

# ==============================
# CORE FUNCTIONS
# ==============================

def print_board(board):
    for i in range(3):
        print(board[i*3:(i+1)*3])
    print()

def actions(board):
    return [i for i in range(9) if board[i] == ' ']

def result(board, action, player):
    new_board = board.copy()
    new_board[action] = player
    return new_board

def player(board):
    return 'X' if board.count('X') == board.count('O') else 'O'

def winner(board):
    lines = [
        [0,1,2],[3,4,5],[6,7,8],
        [0,3,6],[1,4,7],[2,5,8],
        [0,4,8],[2,4,6]
    ]
    for line in lines:
        if board[line[0]] != ' ' and all(board[i] == board[line[0]] for i in line):
            return board[line[0]]
    return None

def terminal(board):
    return winner(board) is not None or ' ' not in board

def utility(board):
    w = winner(board)
    if w == 'X': return 1
    elif w == 'O': return -1
    return 0

# ==============================
# MINIMAX
# ==============================

def max_value(board):
    global nodes_minimax
    nodes_minimax += 1

    if terminal(board):
        return utility(board)

    v = -math.inf
    for a in actions(board):
        v = max(v, min_value(result(board, a, 'X')))
    return v

def min_value(board):
    global nodes_minimax
    nodes_minimax += 1

    if terminal(board):
        return utility(board)

    v = math.inf
    for a in actions(board):
        v = min(v, max_value(result(board, a, 'O')))
    return v

def minimax_decision(board):
    best_move = None
    best_value = -math.inf

    for a in actions(board):
        val = min_value(result(board, a, 'X'))
        if val > best_value:
            best_value = val
            best_move = a

    return best_move

# ==============================
# ALPHA-BETA
# ==============================

def max_ab(board, alpha, beta):
    global nodes_ab
    nodes_ab += 1

    if terminal(board):
        return utility(board)

    v = -math.inf
    for a in actions(board):
        v = max(v, min_ab(result(board, a, 'X'), alpha, beta))
        if v >= beta:
            return v
        alpha = max(alpha, v)
    return v

def min_ab(board, alpha, beta):
    global nodes_ab
    nodes_ab += 1

    if terminal(board):
        return utility(board)

    v = math.inf
    for a in actions(board):
        v = min(v, max_ab(result(board, a, 'O'), alpha, beta))
        if v <= alpha:
            return v
        beta = min(beta, v)
    return v

def alphabeta_decision(board):
    best_move = None
    best_value = -math.inf

    for a in actions(board):
        val = min_ab(result(board, a, 'X'), -math.inf, math.inf)
        if val > best_value:
            best_value = val
            best_move = a

    return best_move

# ==============================
# GAME LOOP (Human vs AI)
# ==============================

def play_game():
    board = [' '] * 9

    while True:
        print_board(board)

        if terminal(board):
            break

        # AI (X)
        move = alphabeta_decision(board)
        print("AI plays at:", move)
        board = result(board, move, 'X')

        if terminal(board):
            break

        print_board(board)

        # Human (O)
        move = int(input("Enter position (0-8): "))
        board = result(board, move, 'O')

    print_board(board)

    w = winner(board)
    if w:
        print(w, "wins")
    else:
        print("Draw")

    print("\nNodes evaluated:")
    print("Minimax:", nodes_minimax)
    print("Alpha-Beta:", nodes_ab)

# Run
play_game()
```
