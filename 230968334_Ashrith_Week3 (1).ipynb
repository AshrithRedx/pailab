## Experiment 1: Breadth-First Search (BFS) for Maze Navigation

Problem Statement:
Consider a maze represented as a graph, where each node corresponds to a location in the maze and
edges represent valid paths between locations.
An agent is placed at a start node and must reach a goal node by traversing the maze.
Task:
1. Formulate the maze navigation problem as a search problem by clearly defining:
o State space
o Initial state
o Actions
o Transition model
o Goal test
o Path cost
2. Implement the Breadth-First Search (BFS) algorithm to find the shortest path from the start
node to the goal node.
3. Print the path returned by BFS.
4. Count and display the number of nodes expanded during the search.
Constraints:
• The maze is unweighted.
• The environment is fully observable, deterministic, static, discrete, and single-agent.
• BFS must be implemented without using any built-in graph search libraries

```python
from collections import deque

def bfs(graph, start, goal):
    # Queue stores paths, not just nodes
    queue = deque()
    queue.append([start])

    # Visited set to avoid revisiting nodes
    visited = set()

    # Counter for expanded nodes
    nodes_expanded = 0

    while queue:
        # Remove first path from queue (FIFO)
        path = queue.popleft()
        current_node = path[-1]

        # Goal test
        if current_node == goal:
            return path, nodes_expanded

        # Expand node if not visited
        if current_node not in visited:
            visited.add(current_node)
            nodes_expanded += 1

            # Generate successor states
            for neighbor in graph[current_node]:
                new_path = path + [neighbor]
                queue.append(new_path)

    # If goal not reachable
    return None, nodes_expanded
```

```python
maze = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E', 'G'],
    'G': []
}

start_node = 'A'
goal_node = 'G'

path, expanded = bfs(maze, start_node, goal_node)

print("Path found:", path)
print("Number of nodes expanded:", expanded)
```

```python
def dfs(graph, current, goal, path, visited, counter):
    visited.add(current)
    counter[0] += 1
    path.append(current)
    
    
    if current == goal:
        return True
    
    
    for neighbor in graph[current]:
        if neighbor not in visited:
            if dfs(graph, neighbor, goal, path, visited, counter):
                return True
    
    
    path.pop()
    return False


path = []
visited = set()
counter = [0]


found = dfs(maze, 'A', 'G', path, visited, counter)


if found:
    print("Path found:", path)
else:
    print("No path found")
    

print("Nodes expanded:", counter[0])
```

## Experiment 3: Solving the 8-Puzzle Problem Using BFS

Problem Statement:
The 8-puzzle problem consists of a 3×3 grid containing eight numbered tiles and one blank space. A
tile adjacent to the blank space can be moved into the blank position.
Given an initial configuration and a goal configuration, implement Breadth-First Search (BFS) to find
a sequence of moves that transforms the initial state into the goal state.
Task Requirements:
1. Represent the puzzle state using a suitable data structure.
2. Define:
o Initial state
o Goal state
o Valid actions (up, down, left, right)
o Transition model
3. Implement BFS to explore the state space.
4. Output:
o The sequence of states from the initial state to the goal state
o The total number of states expanded during the search
Constraints:
• Use BFS to guarantee the shortest solution.
• Avoid revisiting previously explored states.
• Assume the given initial state is solvable.
Expected Outcome:
The algorithm should produce the shortest sequence of moves required to reach the goal
configuration.

```python
from collections import deque


GOAL_STATE = (1,2,3,4,5,6,7,8,0)


moves = {
0: [1,3], 1: [0,2,4], 2: [1,5],
3: [0,4,6], 4: [1,3,5,7], 5: [2,4,8],
6: [3,7], 7: [4,6,8], 8: [5,7]
}


def bfs_8_puzzle(start):
    queue = deque([[start]])
    visited = set([start])
    expanded = 0


    while queue:
        path = queue.popleft()
        state = path[-1]
        expanded += 1


        if state == GOAL_STATE:
            return path, expanded


        zero = state.index(0)
        for move in moves[zero]:
            new_state = list(state)
            new_state[zero], new_state[move] = new_state[move], new_state[zero]
            new_state = tuple(new_state)


            if new_state not in visited:
                visited.add(new_state)
                queue.append(path + [new_state])


    return None, expanded


initial_state = (1,2,3,4,0,6,7,5,8)
solution, expanded = bfs_8_puzzle(initial_state)


print("Steps to goal:", len(solution)-1)
print("States expanded:", expanded)
```

## Experiment 4: Performance Comparison of BFS and DFS

Problem Statement:
Breadth-First Search (BFS) and Depth-First Search (DFS) exhibit different performance characteristics
in terms of time complexity, space complexity, and solution optimality.
Design an experiment to compare BFS and DFS when applied to the same search problem.
Task Requirements:
1. Apply both BFS and DFS to a common problem instance (graph traversal or maze navigation).
2. Instrument both algorithms to record:
o Number of nodes expanded
o Execution time
o Maximum memory used (queue or recursion depth)3. Tabulate the results obtained for BFS and DFS.
4. Analyze and explain the observed differences.
Constraints:
• Use the same start and goal states for both algorithms.
• Run each algorithm under identical conditions.
• Use Python’s time module for execution time measurement.
Expected Outcome:
Students should observe that:
• BFS guarantees the shortest path but uses more memory
• DFS uses less memory but may find non-optimal paths

```python
from collections import deque
import time

def bfs_performance(graph, start, goal):
    queue = deque([[start]])
    visited = set()
    nodes_expanded = 0
    max_queue_size = 1

    while queue:
        max_queue_size = max(max_queue_size, len(queue))
        path = queue.popleft()
        current = path[-1]

        if current == goal:
            return path, nodes_expanded, max_queue_size

        if current not in visited:
            visited.add(current)
            nodes_expanded += 1
            for neighbor in graph[current]:
                queue.append(path + [neighbor])

    return None, nodes_expanded, max_queue_size


start_time = time.time()
bfs_path, bfs_expanded, bfs_memory = bfs_performance(maze, 'A', 'G')
bfs_time = time.time() - start_time

print("BFS Path:", bfs_path)
print("BFS Nodes Expanded:", bfs_expanded)
print("BFS Time:", bfs_time)
print("BFS Max Queue Size (Memory):", bfs_memory)
```

```python
def dfs_performance(graph, current, goal, visited, depth_info):
    visited.add(current)
    depth_info["nodes"] += 1
    depth_info["max_depth"] = max(depth_info["max_depth"], len(visited))

    if current == goal:
        return True

    for neighbor in graph[current]:
        if neighbor not in visited:
            if dfs_performance(graph, neighbor, goal, visited, depth_info):
                return True
    return False


depth_info = {"nodes": 0, "max_depth": 0}

start_time = time.time()
visited = set()
dfs_performance(maze, 'A', 'G', visited, depth_info)
dfs_time = time.time() - start_time

print("DFS Nodes Expanded:", depth_info["nodes"])
print("DFS Time:", dfs_time)
print("DFS Max Recursion Depth (Memory):", depth_info["max_depth"])
```

```python
print("\nAlgorithm | Nodes Expanded | Time (s) | Memory Used")
print("--------------------------------------------------")
print(f"BFS       | {bfs_expanded:<14} | {bfs_time:.6f} | Queue Size = {bfs_memory}")
print(f"DFS       | {depth_info['nodes']:<14} | {dfs_time:.6f} | Depth = {depth_info['max_depth']}")
```

### Analysis

Breadth-First Search (BFS) explores nodes level by level, which guarantees the shortest path but requires more memory due to storing multiple paths in the queue.

Depth-First Search (DFS) explores deep paths first and uses less memory, but it does not guarantee an optimal solution.

From the results, BFS expanded more nodes and used more memory, while DFS used less memory but may produce a non-optimal path.
